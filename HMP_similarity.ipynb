The following command is for performing hydrogen mass repartitioning and saved in GROMACS format.


```python
import parmed as pmd
from parmed import tools
amber = pmd.load_file('/content/solvated.prmtop', '/content/solvated.inpcrd')
pmd.tools.actions.HMassRepartition(amber)
amber.save('gromacs.top')
amber.save('gromacs.gro')
amber.save('solvated_hms.prmtop')
amber.save('solvated_hms.inpcrd')
```

```python
import difflib

with open('/content/solvated2.txt', 'r') as f1, open('/content/solvated_hms.txt', 'r') as f2:
    diff = difflib.unified_diff(f1.readlines(), f2.readlines(), fromfile='file1.txt', tofile='/content/solvated_hms.txt')
    num_of_diffs = sum(1 for line in diff if line.startswith('+') or line.startswith('-'))
    print("Jumlah perbedaan antara kedua file: ", num_of_diffs)
```
