```python
import pandas as pd
from docx import Document
from docx.shared import Pt
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT

# Mengunggah file Excel
from google.colab import files
uploaded = files.upload()

# Baca file Excel
file_name = '/content/gcolab_knkhrb.xlsx'  # Ganti dengan nama file Anda
df = pd.read_excel(file_name)

# Membuat dokumen Word
doc = Document()
doc.styles['Normal'].font.name = 'Calibri'

# Tambahkan setiap baris dengan judul bold dan nama penulis, fakultas
for i, row in df.iterrows():
    judul = row['Judul Penelitian']
    nama = row['Nama Penulis']
    fakultas = row['Fakultas']

    # Tambahkan judul dengan format bold, font Calibri, size 12
    judul_paragraph = doc.add_paragraph()
    judul_run = judul_paragraph.add_run(judul)
    judul_run.bold = True
    judul_run.font.size = Pt(12)
    judul_paragraph.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT

    # Tambahkan nama penulis
    nama_fakultas = doc.add_paragraph()
    nama_fakultas.add_run(nama)
    nama_fakultas.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT

    # Tambahkan tulisan "Fakultas" di depan nama fakultas
    nama_fakultas.add_run(f", Fakultas {fakultas}")
    nama_fakultas.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT

# Simpan dokumen Word
doc.save('output.docx')
files.download('output.docx')
```
